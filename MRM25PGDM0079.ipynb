```python
# ===========================
# 0. Environment Setup
# ===========================
!pip install openpyxl mlxtend pmdarima plotly pycaret --quiet
!pip install --upgrade pandas matplotlib


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
from mlxtend.frequent_patterns import apriori, association_rules
from pmdarima import auto_arima
import statsmodels.api as sm
import datetime
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline

# If using Google Drive uncomment below:
# from google.colab import drive
# drive.mount('/content/drive')

# ===========================
# 1. Data Import & Initial Exploration
# ===========================
# Adjust the file_path to point to your 'Online Retail.xlsx' file location
file_path = '/content/FINAL MRM DATASET.xlsx'  # e.g., 'drive/MyDrive/Online Retail.xlsx'
df = pd.read_excel(file_path)

print("---- First 5 Rows ----")
display(df.head())

print("---- Data Info ----")
df.info()

print("---- Data Description (Numeric Columns) ----")
display(df.describe())

print("---- Missing Values ----")
print(df.isnull().sum())

# ===========================
# 2. Data Cleaning & Preparation
# ===========================
# Drop rows with missing CustomerID
df.dropna(subset=['CustomerID'], inplace=True)

# Convert InvoiceDate to datetime if needed
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

# Remove rows with non-positive Quantity or UnitPrice (i.e., returns or errors)
df = df[(df['Quantity'] > 0) & (df['UnitPrice'] > 0)]

# Identify cancellations (InvoiceNo starting with 'C')
cancelled_df = df[df['InvoiceNo'].astype(str).str.startswith('C')]
print("Cancelled/Returned Transactions Count:", len(cancelled_df))

# Remove cancelled invoices to focus on completed sales
df = df[~df['InvoiceNo'].astype(str).str.startswith('C')]

# Create TotalPrice column
df['TotalPrice'] = df['Quantity'] * df['UnitPrice']

# Verify that TotalPrice exists
print("Columns after cleaning:")
print(df.columns)
display(df.head())

# ===========================
# 3. Advanced Exploratory Data Analysis (EDA)
# ===========================

## 3.1 Univariate Analysis
plt.figure(figsize=(8,5))
sns.histplot(df['TotalPrice'], bins=50, kde=True)
plt.title('Distribution of TotalPrice')
plt.show()

plt.figure(figsize=(8,5))
sns.histplot(df['Quantity'], bins=50, kde=True)
plt.title('Distribution of Quantity')
plt.show()

plt.figure(figsize=(8,5))
sns.histplot(df['UnitPrice'], bins=50, kde=True)
plt.title('Distribution of UnitPrice')
plt.show()

## 3.2 Bivariate Analysis
corr_matrix = df[['Quantity', 'UnitPrice', 'TotalPrice']].corr()
plt.figure(figsize=(5,4))
sns.heatmap(corr_matrix, annot=True, cmap='Blues')
plt.title('Correlation Matrix')
plt.show()

plt.figure(figsize=(6,4))
sns.scatterplot(x='Quantity', y='UnitPrice', data=df)
plt.title('Quantity vs. UnitPrice')
plt.show()

plt.figure(figsize=(6,4))
sns.scatterplot(x='Quantity', y='TotalPrice', data=df)
plt.title('Quantity vs. TotalPrice')
plt.show()

## 3.3 Time-Based Analysis
# Set InvoiceDate as index for time-based resampling
df.set_index('InvoiceDate', inplace=True)
daily_revenue = df['TotalPrice'].resample('D').sum().fillna(0)

plt.figure(figsize=(12,6))
plt.plot(daily_revenue.index, daily_revenue.values, color='blue')
plt.title('Daily Revenue Over Time')
plt.xlabel('Date')
plt.ylabel('Revenue')
plt.show()

monthly_revenue = df['TotalPrice'].resample('M').sum().fillna(0)
plt.figure(figsize=(12,6))
plt.plot(monthly_revenue.index, monthly_revenue.values, marker='o', color='green')
plt.title('Monthly Revenue Over Time')
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.show()

# Reset index for further analyses
df.reset_index(inplace=True)

## 3.4 Country-Level Analysis
country_revenue = df.groupby('Country')['TotalPrice'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(8,5))
sns.barplot(x=country_revenue.values, y=country_revenue.index)
plt.title('Top 10 Countries by Revenue')
plt.xlabel('Revenue')
plt.ylabel('Country')
plt.show()

## 3.5 Product-Level Analysis
product_revenue = df.groupby('Description')['TotalPrice'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(8,5))
sns.barplot(x=product_revenue.values, y=product_revenue.index)
plt.title('Top 10 Products by Revenue')
plt.xlabel('Revenue')
plt.ylabel('Product Description')
plt.show()

product_quantity = df.groupby('Description')['Quantity'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(8,5))
sns.barplot(x=product_quantity.values, y=product_quantity.index)
plt.title('Top 10 Products by Quantity Sold')
plt.xlabel('Quantity Sold')
plt.ylabel('Product Description')
plt.show()

# ===========================
# 4. RFM Analysis (Customer Segmentation)
# ===========================
# Define snapshot date (1 day after last invoice date)
snapshot_date = df['InvoiceDate'].max() + datetime.timedelta(days=1)

rfm_df = df.groupby('CustomerID').agg({
    'InvoiceDate': lambda x: (snapshot_date - x.max()).days,  # Recency
    'InvoiceNo': 'nunique',                                   # Frequency
    'TotalPrice': 'sum'                                       # Monetary
}).reset_index()

rfm_df.columns = ['CustomerID', 'Recency', 'Frequency', 'Monetary']

# Calculate quantiles for RFM scoring
quantiles = rfm_df.quantile(q=[0.25, 0.5, 0.75])
def rfm_score(x, col):
    if x <= quantiles[col][0.25]:
        return 4
    elif x <= quantiles[col][0.50]:
        return 3
    elif x <= quantiles[col][0.75]:
        return 2
    else:
        return 1

rfm_df['R_Score'] = rfm_df['Recency'].apply(lambda x: rfm_score(x, 'Recency'))
rfm_df['F_Score'] = rfm_df['Frequency'].apply(lambda x: rfm_score(x, 'Frequency'))
rfm_df['M_Score'] = rfm_df['Monetary'].apply(lambda x: rfm_score(x, 'Monetary'))

rfm_df['RFM_Segment'] = rfm_df['R_Score'].astype(str) + rfm_df['F_Score'].astype(str) + rfm_df['M_Score'].astype(str)
rfm_df['RFM_Score'] = rfm_df[['R_Score','F_Score','M_Score']].sum(axis=1)

print("---- RFM Analysis Sample ----")
display(rfm_df.head(10))

plt.figure(figsize=(8,5))
sns.countplot(x=rfm_df['RFM_Score'])
plt.title('Distribution of RFM Scores')
plt.xlabel('RFM Score')
plt.ylabel('Count of Customers')
plt.show()

best_customers = rfm_df[rfm_df['RFM_Score'] >= 10]
print("Number of Best Customers (RFM Score >= 10):", len(best_customers))

# ===========================
# 5. Clustering-Based Segmentation (Optional)
# ===========================
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA

features = rfm_df[['Recency', 'Frequency', 'Monetary']].copy()
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

sse = []
for k in range(2, 10):
    km = KMeans(n_clusters=k, random_state=42)
    km.fit(scaled_features)
    sse.append(km.inertia_)

plt.figure(figsize=(6,4))
plt.plot(range(2,10), sse, marker='o')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('SSE')
plt.show()

# Using k=4 as an example
kmeans = KMeans(n_clusters=4, random_state=42)
rfm_df['Cluster'] = kmeans.fit_predict(scaled_features)

pca = PCA(n_components=2)
pca_features = pca.fit_transform(scaled_features)

plt.figure(figsize=(6,5))
plt.scatter(pca_features[:,0], pca_features[:,1], c=rfm_df['Cluster'], cmap='viridis')
plt.title('Customer Clusters (PCA 2D Projection)')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.colorbar()
plt.show()

cluster_analysis = rfm_df.groupby('Cluster').agg({
    'Recency': 'mean',
    'Frequency': 'mean',
    'Monetary': 'mean',
    'CustomerID': 'count'
}).rename(columns={'CustomerID': 'Count'}).reset_index()
print("---- Cluster Analysis ----")
display(cluster_analysis)

# ===========================
# 6. Market Basket Analysis (Association Rules)
# ===========================
# Focus on a specific country, e.g., 'United Kingdom'
country_data = df[df['Country'] == 'United Kingdom'].copy()

basket = (country_data.groupby(['InvoiceNo', 'Description'])['Quantity']
          .sum().unstack().fillna(0))
basket = basket.applymap(lambda x: 1 if x > 0 else 0)

freq_itemsets = apriori(basket, min_support=0.01, use_colnames=True)
freq_itemsets.sort_values('support', ascending=False, inplace=True)
print("---- Frequent Itemsets ----")
display(freq_itemsets.head(10))

rules = association_rules(freq_itemsets, metric='lift', min_threshold=1)
rules.sort_values('confidence', ascending=False, inplace=True)
print("---- Association Rules ----")
display(rules.head(10))

filtered_rules = rules[(rules['confidence'] > 0.5) & (rules['lift'] > 1.2)]
print("---- Filtered Association Rules (Confidence > 0.5 & Lift > 1.2) ----")
display(filtered_rules.head(10))

# ===========================
# 7. Advanced Time Series Forecasting (Daily Sales)
# ===========================
df.set_index('InvoiceDate', inplace=True)
daily_sales = df['TotalPrice'].resample('D').sum().fillna(0)

plt.figure(figsize=(12,6))
plt.plot(daily_sales.index, daily_sales.values, label='Daily Sales')
plt.title('Daily Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.show()

# Train-Test split
train_size = int(len(daily_sales) * 0.8)
train_data = daily_sales[:train_size]
test_data = daily_sales[train_size:]

model = auto_arima(train_data, seasonal=False, m=1, trace=False, error_action='ignore')
model.fit(train_data)
forecast_period = len(test_data)
forecast = model.predict(n_periods=forecast_period)

plt.figure(figsize=(12,6))
plt.plot(train_data.index, train_data, label='Train')
plt.plot(test_data.index, test_data, label='Test')
plt.plot(test_data.index, forecast, label='Forecast', color='red')
plt.title('Time Series Forecasting: Daily Sales')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.show()

df.reset_index(inplace=True)  # Reset index back

# ===========================
# 8. Price Elasticity Analysis (Regression Approach)
# ===========================
price_elasticity_df = df.groupby('StockCode').agg({
    'Quantity': 'sum',
    'UnitPrice': 'mean'
}).reset_index()

corr_val = price_elasticity_df[['Quantity', 'UnitPrice']].corr().iloc[0,1]
print(f"Correlation between Average Price and Total Quantity Sold: {corr_val:.3f}")

# Use log-log regression to estimate price elasticity
X = price_elasticity_df[['UnitPrice']]
y = price_elasticity_df['Quantity']
X_log = np.log(X)
y_log = np.log(y)
X_log = sm.add_constant(X_log)
elasticity_model = sm.OLS(y_log, X_log).fit()
print("---- Price Elasticity Regression Summary ----")
print(elasticity_model.summary())

# ===========================
# 9. Cancellation/Return Analysis
# ===========================
df['IsCancelled'] = df['InvoiceNo'].astype(str).str.startswith('C')
cancel_rates = df.groupby('StockCode')['IsCancelled'].mean().sort_values(ascending=False)
print("---- Top 10 StockCodes by Cancellation Rate ----")
print(cancel_rates.head(10))

# ===========================
# 10. Customer Lifetime Value (CLV) Analysis (Simplified)
# ===========================
clv_df = df.groupby('CustomerID')['TotalPrice'].sum().reset_index().rename(columns={'TotalPrice':'CLV'})
clv_df.sort_values('CLV', ascending=False, inplace=True)
print("---- Top 10 Customers by CLV ----")
display(clv_df.head(10))

# ===========================
# 11. Anomaly Detection (Daily Sales)
# ===========================
df.set_index('InvoiceDate', inplace=True)
daily_sales = df['TotalPrice'].resample('D').sum().fillna(0)
rolling_mean = daily_sales.rolling(window=7).mean()
rolling_std = daily_sales.rolling(window=7).std()
upper_bound = rolling_mean + 3 * rolling_std
lower_bound = rolling_mean - 3 * rolling_std
anomalies = daily_sales[(daily_sales > upper_bound) | (daily_sales < lower_bound)]
print("---- Anomalies in Daily Sales ----")
display(anomalies)

# Reset index at the end
df.reset_index(inplace=True)

print("----- Analysis Completed -----")
```

```python
# ===========================
# Snippet 2: Data Import & Initial Exploration (Deep)
# ===========================
file_path = '/content/Online Retail.xlsx'  # Adjust path as needed
df = pd.read_excel('/content/FINAL MRM DATASET.xlsx')

# Optionally work with a sample for faster development:
df = df.sample(frac=0.1, random_state=42)

# Memory usage of dataframe
print("Memory Usage (before optimization):")
print(df.memory_usage(deep=True))

# Display first few rows
print("---- First 5 Rows ----")
display(df.head())

# Dataframe Info and Description
print("---- Data Info ----")
df.info()

print("---- Numeric Description ----")
display(df.describe())

print("---- Categorical Description ----")
display(df.describe(include=['object']))

print("---- Missing Values ----")
print(df.isnull().sum())

# Plot distributions for key columns (density plots)
plt.figure(figsize=(8,4))
sns.kdeplot(df['Quantity'], shade=True)
plt.title('Density of Quantity')
plt.show()

plt.figure(figsize=(8,4))
sns.kdeplot(df['UnitPrice'], shade=True)
plt.title('Density of UnitPrice')
plt.show()

# Memory usage after potential type conversion can be added later in cleaning.
```

```python
# ===========================
# Snippet 3: Data Cleaning & Preparation (Deep)
# ===========================
# Drop rows missing CustomerID
df.dropna(subset=['CustomerID'], inplace=True)

# Convert InvoiceDate to datetime
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'], errors='coerce')

# Remove rows with non-positive Quantity or UnitPrice (e.g., returns/errors)
df = df[(df['Quantity'] > 0) & (df['UnitPrice'] > 0)]

# Optionally, remove extreme outliers using IQR (for UnitPrice and Quantity)
def remove_outliers(data, col):
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return data[(data[col] >= lower_bound) & (data[col] <= upper_bound)]

df = remove_outliers(df, 'UnitPrice')
df = remove_outliers(df, 'Quantity')

# Identify cancellations (InvoiceNo starting with 'C')
cancelled_df = df[df['InvoiceNo'].astype(str).str.startswith('C')]
print("Cancelled/Returned Transactions Count:", len(cancelled_df))

# Remove cancelled invoices
df = df[~df['InvoiceNo'].astype(str).str.startswith('C')]

# Create TotalPrice column
df['TotalPrice'] = df['Quantity'] * df['UnitPrice']

# Optimize data types: CustomerID to string, downcast numeric types
df['CustomerID'] = df['CustomerID'].astype(str)
df['Quantity'] = pd.to_numeric(df['Quantity'], downcast='integer')
df['UnitPrice'] = pd.to_numeric(df['UnitPrice'], downcast='float')
df['TotalPrice'] = pd.to_numeric(df['TotalPrice'], downcast='float')

print("Columns after cleaning and optimization:")
print(df.columns)
display(df.head())

# Memory usage after cleaning
print("Memory Usage (after cleaning):")
print(df.memory_usage(deep=True))
gc.collect()
```
